3. Write a fun which will take input as a dict give me out as a list of all values even in case of 2 level nesting it should work

```python
def value(a):
    """ fun to display all the dict values""" 
    l=[]
    if type(a)==dict:
        for i in a.values():
            if type(i)!=dict:
                l.append(i)
            else:
                for j in i.values():
                    l.append(j)
        return l
                
        
```

5. write a fun which takes list as an input and give me concatination of all element as  output. 

```python
def concat(*args):
    """ fun to take n number of lists and give concatenaaton as one list"""
    l=[]
    for i in args:
        if type(i)==list:
            l=l+i
    return l
    
    
```
